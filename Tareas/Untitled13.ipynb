```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random


```

```python


def cargarDatos():
    datasetList = ["taxis", "flights", "planets"]
    rand_idx = random.randrange(len(datasetList))
    random_dataset = datasetList[rand_idx]
    print("El dataset que utilizarás es: ", random_dataset)
    df = sns.load_dataset(random_dataset)
    return df

```

```python
# Asigna el dataset al azar y carga el dataset
df = cargarDatos()
df.head()


```

```python
#Resumen del dataset
df.info()
```

```python
# Assuming 'month' is the categorical column
if 'month' in df.columns and df['month'].dtype.name == 'category':
    # Group by month and sum passengers
    monthly_passengers = df.groupby('month')['passengers'].sum().reset_index()

    # Create a bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='month', y='passengers', data=monthly_passengers)
    plt.title('Total de Pasajeros por Mes')
    plt.xlabel('Mes')
    plt.ylabel('Total de Pasajeros')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No se encontró una columna categórica llamada 'month' en el DataFrame.")
```

```python
# prompt: Analiza de manera gráfica un dato cuantitativo de tu dataset.

import matplotlib.pyplot as plt
# Seleccionar una columna cuantitativa (asegúrate de que exista en el DataFrame)
# Puedes imprimir df.info() para ver las columnas y sus tipos de datos.
# Reemplaza 'nombre_columna_cuantitativa' con el nombre real de una columna cuantitativa en tu dataset.

# Load the dataset
df = cargarDatos()

columna_cuantitativa = None
for col in df.columns:
  if df[col].dtype in ['int64', 'float64']:
    columna_cuantitativa = col
    break

if columna_cuantitativa is not None:
  print(f"Analizando la columna cuantitativa: {columna_cuantitativa}")

  # Histograma para visualizar la distribución
  plt.figure(figsize=(10, 6))
  sns.histplot(data=df, x=columna_cuantitativa, kde=True)
  plt.title(f'Distribución de {columna_cuantitativa}')
  plt.xlabel(columna_cuantitativa)
  plt.ylabel('Frecuencia')
  plt.show()

  # Boxplot para visualizar la dispersión y valores atípicos
  plt.figure(figsize=(10, 6))
  sns.boxplot(data=df, y=columna_cuantitativa)
  plt.title(f'Boxplot de {columna_cuantitativa}')
  plt.ylabel(columna_cuantitativa)
  plt.show()

else:
  print("No se encontró una columna cuantitativa en el DataFrame para analizar.")
```

```python
# prompt: Método gráfico multivariado para datos cuantitativos

import matplotlib.pyplot as plt
# Gráfico de dispersión (Scatter Plot) para visualizar la relación entre dos columnas cuantitativas
# Debes identificar dos columnas cuantitativas en tu dataset para este gráfico.
# Reemplaza 'columna_cuantitativa_1' y 'columna_cuantitativa_2' con los nombres reales.

cuantitative_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]

if len(cuantitative_cols) >= 2:
  columna_cuantitativa_1 = cuantitative_cols[0]
  columna_cuantitativa_2 = cuantitative_cols[1]

  print(f"Visualizando la relación entre {columna_cuantitativa_1} y {columna_cuantitativa_2}")

  plt.figure(figsize=(10, 6))
  sns.scatterplot(data=df, x=columna_cuantitativa_1, y=columna_cuantitativa_2)
  plt.title(f'Relación entre {columna_cuantitativa_1} y {columna_cuantitativa_2}')
  plt.xlabel(columna_cuantitativa_1)
  plt.ylabel(columna_cuantitativa_2)
  plt.show()
else:
  print("No hay suficientes columnas cuantitativas para crear un gráfico de dispersión.")

# Gráfico de pares (Pair Plot) para visualizar las relaciones entre múltiples columnas cuantitativas
# Esto puede ser útil para explorar correlaciones entre varias variables.
# Puede tomar tiempo en datasets grandes.

if len(cuantitative_cols) >= 2:
  print("Generando gráfico de pares para columnas cuantitativas...")
  # Seleccionar un subconjunto de columnas cuantitativas si hay muchas para evitar sobrecarga
  num_cols_for_pairplot = min(len(cuantitative_cols), 5) # Limita a 5 columnas para el ejemplo
  cols_for_pairplot = cuantitative_cols[:num_cols_for_pairplot]

  sns.pairplot(df[cols_for_pairplot])
  plt.suptitle('Gráfico de Pares de Columnas Cuantitativas Seleccionadas', y=1.02) # Título general
  plt.show()
else:
    print("No hay suficientes columnas cuantitativas para crear un gráfico de pares.")
```
